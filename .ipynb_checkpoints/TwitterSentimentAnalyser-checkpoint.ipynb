```python
# We'll download and use Niek Sanders's Tweet Sentiment Corpus (csv file). 
def createTrainingCorpus(corpusFile, tweetDataFile):
    corpus=[]
    
    # Read the downloaded csv file to store the tweet id (as twitter doesn't directly gives posts), tweet 
    # label(positive/negative/neutral) and the tweet topic
    with open(corpusFile, 'rt') as csvFile:
        lineReader = csv.reader(csvFile, delimiter=',', quotechar="\"")
        for row in lineReader:
            corpus.append({"tweet_id":row[2], "label":row[1], "topic":row[0]})
    
    # Twitter has rate limit of 180 requests per window(15 mins) for tweet searches
    rate_limit = 180
    sleep_time = 900/180
    
    trainingData=[]
    for tweet in corpus:
        # Iterate over the corpus to get the actual text from the tweet Id
        try:
            status=api.GetStatus(tweet["tweet_id"])    # return status object
            print("Tweet fetched : ", status.text)
            # Adding 'text' label to the dictionary tweet
            tweet["text"]=status.text
            trainingData.append(tweet)
            time.sleep(sleep_time)    # to avoid being rate limited
        except Exception as ex:
            print("Error occurred for tweet Id %s : %s" %(str(tweet["tweet_id"]), str(ex)))
    
    # Now we store the downloaded tweets in a file
    with open(tweetDataFile, 'wt') as csvFile:
        lineWriter = csv.writer(csvFile, delimiter=',', quotechar="\"")
        for tweet in trainingData:
            try:
                lineWriter.writerow([tweet["tweet_id"],tweet["text"],tweet["label"],tweet["topic"]])
            except Exception as ex:
                print("Error occurred in writing file : " + str(ex))
    
    return trainingData        
```

```python
def createLimitedTrainingCorpus(corpusFile, tweetDataFile):
    corpus=[]
    
    # Read the downloaded csv file to store the tweet id (as twitter doesn't directly gives posts), tweet 
    # label(positive/negative/neutral) and the tweet topic
    with open(corpusFile, 'rt') as csvFile:
        lineReader = csv.reader(csvFile, delimiter=',', quotechar="\"")
        for row in lineReader:
            corpus.append({"tweet_id":row[2], "label":row[1], "topic":row[0]})
    
    trainingData=[]
    for label in ["positive", "negative"]:
        i=1
        for tweet in corpus:
            # This time we'll only download 50 tweets from each of the labels
            if tweet["label"]==label and i<=50:
                try:
                    status=api.GetStatus(tweet["tweet_id"])
                    print("Tweet fetched : ", status.text)
                    # Adding 'text' label to the dictionary tweet
                    tweet["text"]=status.text
                    trainingData.append(tweet)
                    i = i+1
                except Exception as ex:
                    print("Error occurred for tweet Id %s : %s" %(str(tweet["tweet_id"]), str(ex)))
    
    # Now we store the downloaded tweets in a file
    with open(tweetDataFile, 'wt') as csvFile:
        lineWriter = csv.writer(csvFile, delimiter=',', quotechar="\"")
        for tweet in trainingData:
            try:
                lineWriter.writerow([tweet["tweet_id"],tweet["text"],tweet["label"],tweet["topic"]])
            except Exception as ex:
                print("Error occurred in writing file : " + str(ex))
                    
    return trainingData
```

```python
# Importing libraries to start pre-processing

import re
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.corpus import stopwords
```

```python
class PreProcessTweets:
    """
    This class has functions to clean up the text extracted from the twitter Api.
    
    """
    def __init__(self):
        # to get all the unwanted characters and words together
        self._stopwords=set(stopwords.words('english')+list(punctuation)+['AT_USER', 'URL'])
    
    def processTweets(self, listOfTweets):
        # 'listOfTweets' is dictionary with keys : 'text' and 'label'
        processedTweets=[]
            
        for tweet in listOfTweets:
            # list of tuple (list of words in tweet, label of tweet)
            processedTweets.append((self._processTweet(tweet["text"]), tweet["label"]))
        return processedTweets
    
    def _processTweet(self, tweet):
        # to lowercase
        tweet=tweet.lower()
        
        # Replace links with the word 'URL'
        tweet=re.sub('((www\.[^\s]+)|(https?://[^\s]+))', 'URL', tweet)
        
        # Replace @username with 'AT_USER'
        tweet=re.sub('@[^\s]+', 'AT_USER', tweet)
        
        # Replace #word with word
        tweet=re.sub(r'#([^\s]+)', r'\1', tweet)
        
        # Correct repeating words - Ex 'Haaaaaaapppy' to 'Haappy'
        tweet=re.sub(r'(.)\1+', r'\1\1', tweet) 
        
        # tokenize the tweet in a list of words
        tweet=word_tokenize(tweet)
        
        # return the list - stopwords
        return [word for word in tweet if word not in self._stopwords and word.isalpha()]
```

```python
# Importing libraries to start classification task

import nltk
```

```python
def buildVocabulary(ppTrainingData):
    all_words=[]
    
    # this will catch all the words in the corpus together in one list
    for (words, sentiment) in ppTrainingData:
        all_words.extend(words)
    
    # creating a dictionary key : word, value : frequency
    wordList=nltk.FreqDist(all_words)
    
    #return the list of all the unique words in the vocab
    word_Features=wordList.keys()
    
    return word_Features
```

```python
# NLTK has an apply_features function that takes in a user-defined function to extract features 
# from training data. We want to define our extract features function to take each tweet in 
# The training data and represent it with the presence or absence of a word in the vocabulary 

def extract_Features(tweet):
    tweet_words=set(tweet)
    features={}
    for word in word_Features:
        features['contains(%s)' %word]=(word in tweet_words)
        # This will give us a dictionary , with keys like 'contains word1' and 'contains word2'
        # and values as True or False 
    return features
```

```python
# Now we can extract the features and train the classifier
word_Features=buildVocabulary(pptrainingData)
trainingFeatures=nltk.classify.apply_features(extract_Features, pptrainingData)
# apply_features will take the extract_features function we defined above, and apply it to each element of
# ppTrainingData. It automatically identifies that each of those elements is actually a tuple, so it takes
# the first element of the tuple to be the text and second element to be the label, and applies the 
# function only on the text 
```

```python
# Support Vector Machines
from nltk.corpus import sentiwordnet as swn
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

# We have to change the form of the data slightly. SKLearn has a CountVectorizer object. 
# It will take in documents and directly return a term-document matrix with the frequencies of 
# a word in the document. It builds the vocabulary by itself. We will give the trainingData 
# and the labels separately to the SVM classifier and not as tuples. 
# Another thing to take care of, if you built the Naive Bayes for more than 2 classes, 
# SVM can only classify into 2 classes - it is a binary classifier. 
```

```python
# Create sentences out of list of words
svmTrainingData=[' '.join(tweet[0]) for tweet in pptrainingData]
```

```python
vectorizer = CountVectorizer(min_df=1)
X = vectorizer.fit_transform(svmTrainingData).toarray()
vocabulary=vectorizer.get_feature_names()
```

```python
print(vocabulary[5:15])
print(X.shape)
print(len(vocabulary))
```

```python
# using sentiwordnet to add some weights to the features 

swn_weights=[]

for word in vocabulary:
    try:
        # this code is put in try block as all words might not be found in the sentiwordnet
        synset=list(swn.senti_synsets(word))
        # first synset used for scoring as this represents the most common usage of the word. We can also 
        # use the average of all the synsets
        common_meaning=synset[0]
        # print(word, common_meaning)
        if common_meaning.pos_score()>common_meaning.neg_score():
            # for greater pos_score it is used as weight
            weight=common_meaning.pos_score()
        elif common_meaning.pos_score()<common_meaning.neg_score():
            # for greater neg_score use negative of that as weight
            weight=-common_meaning.neg_score()
        else:
            weight=0
    except Exception as ex:
        # We expect some errors here. Ex - if there are no synsets returned. Ex - 'ipad', 'iphone'. For whatever
        # reason the error is, we set the weight to 0
        weight=0
        print("Error in calculating word weights for '"+word+"': "+str(ex))
        
    swn_weights.append(weight)
```

```python
swn_X=[]
for row in X:
    swn_X.append(np.multiply(row, np.array(swn_weights)))
# Converting the list to a vertical numpy array
swn_X = np.vstack(swn_X)
```

```python
labels_To_array={"positive":1, "negative":2}
labels=[labels_To_array[tweet[1]] for tweet in pptrainingData]
y=np.array(labels)
```

```python
# Building the SVM classifier
from sklearn.svm import SVC
SVMClassifier=SVC()
SVMClassifier.fit(swn_X,y)
```

```python
# Running the two classifiers on the 100 downloaded tweets

# Naive bayes 
NBResultLabels=[NBayesClassifier.classify(extract_Features(tweet[0])) for tweet in pptestData]

# SVM
SVMResultLabels=[]
for tweet in pptestData:
    tweet_sentence=' '.join(tweet[0])
    svmFeatures=np.multiply(vectorizer.transform([tweet_sentence]).toarray(), np.array(swn_weights))
    # svmFeatures.shape = (1, 504)
    SVMResultLabels.append(SVMClassifier.predict(svmFeatures)[0])  # predict returns an array
```
